```python
from langsmith import Client
from langchain.smith import RunEvalConfig, run_on_dataset
from langchain.embeddings import HuggingFaceEmbeddings
from langchain import chat_models, prompts, smith
from langchain.schema import output_parser


# Define your runnable or chain below.
prompt = prompts.ChatPromptTemplate.from_messages(
  [
    ("system", "You are a helpful AI Assistant in evaluating chatbot responses. Testing data is comprised of a chatbot. The Chatbot was assigned task to act as personal medical assistant for the patients"),
    ("human", "{question}")
  ]
)
llm = chat_models.ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
chain = prompt | llm | output_parser.StrOutputParser()

evaluation_config = RunEvalConfig(
    evaluators=[
        # You can define an arbitrary criterion as a key: value pair in the criteria dict
        #RunEvalConfig.Criteria("cot_qa"),
        # We provide some simple default criteria like "conciseness" you can use as well
        "qa",
        "context_qa",
        "embedding_distance",

        RunEvalConfig.Criteria("conciseness"),
        #RunEvalConfig.LabeledCriteria("relevance"),
        RunEvalConfig.Criteria("coherence"),
        #RunEvalConfig.LabeledCriteria("harmfulness"),
        RunEvalConfig.Criteria("helpfulness"),
        
    ],
    custom_evaluators=[],
    eval_llm=ChatOpenAI(model=llm_name, temperature=0)
)

client = Client()
run_on_dataset(
   dataset_name="Our_Chatbot_Data",
    #llm_or_chain_factory= LLM | constructor for chain | agent,
    llm_or_chain_factory=chain,
    #chain = prompt | llm ,
    evaluation=evaluation_config,
    #project_name="test-monthly-depot-99",
    concurrency_level=5,
    verbose=True,
    client=client,
)
```
